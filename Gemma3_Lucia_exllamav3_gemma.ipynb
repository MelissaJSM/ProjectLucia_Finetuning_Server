# 📊 LLM 모델별 용량 및 필요 VRAM 추정표

> **참고:**
> * **Lock / Unlock:** 두 버전의 용량은 동일하므로 하나로 통합하여 표기했습니다.
> * **추정 VRAM:** 모델 가중치(파일 크기) + `2~3GB` (KV Cache 및 시스템 오버헤드)를 포함한 수치입니다. 컨텍스트 길이(Context Length)를 길게 설정하면 더 많은 메모리가 필요합니다.

## 1. High-End (24GB VRAM 이상 권장)
**Gemma-4 31B / 26B-A4B** 및 **Gemma-3 27B**와 같은 대형 고성능 모델들이 속하며, 4bit 이상의 양자화 모델을 원활하게 구동하려면 24GB 이상의 VRAM을 갖춘 GPU가 필요합니다.

| 모델 (Model) | 양자화 (Bit) | 파일 용량 | 추정 필요 VRAM | 권장 GPU |
| :--- | :---: | :---: | :---: | :--- |
| **Gemma-4 31B** | **8 bit** | 31.57 GB | **~34.5 GB** | A6000 / Mac Studio 등 |
| **Gemma-3 27B** | **8 bit** | 27.50 GB | **~30.5 GB** | A6000 / 3090 NVLink |
| **Gemma-4 26B-A4B** | **8 bit** | 26.48 GB | **~29.5 GB** | A6000 / 3090 NVLink |
| **Gemma-4 31B** | **6 bit** | 24.73 GB | **~27.5 GB** | A6000 / 3090 NVLink |
| **Gemma-3 27B** | **6 bit** | 21.54 GB | **~24.5 GB** | RTX 3090 / 4090 (여유 없음) |
| **Gemma-4 26B-A4B** | **6 bit** | 20.74 GB | **~23.5 GB** | RTX 3090 / 4090 (여유 없음) |
| **Gemma-4 31B** | **4 bit** | 17.89 GB | **~20.9 GB** | RTX 3090 / 4090 |
| **Gemma-3 27B** | **4 bit** | 15.58 GB | **~18.6 GB** | RTX 3090 / 4090 |
| **Gemma-4 26B-A4B** | **4 bit** | 15.00 GB | **~18.0 GB** | RTX 3090 / 4090 |

---

## 2. Sweet Spot (12GB ~ 16GB VRAM 권장)
**Gemma-3 12B** 및 고용량 모델의 **2bit 극한 양자화 버전**이 위치하며, 일반적인 하이엔드 게이밍 GPU에서 가장 효율이 좋은 구간입니다.

| 모델 (Model) | 양자화 (Bit) | 파일 용량 | 추정 필요 VRAM | 권장 GPU |
| :--- | :---: | :---: | :---: | :--- |
| **Gemma-3 12B** | **8 bit** | 12.81 GB | **~14.8 GB** | RTX 4080 / 4070 Ti Super |
| **Gemma-4 31B** | **2 bit** | 11.04 GB | **~13.5 GB** | RTX 4080 (16GB) |
| **Gemma-3 27B** | **2 bit** | 9.62 GB | **~12.6 GB** | RTX 4080 (16GB) |
| **Gemma-3 12B** | **6 bit** | 10.31 GB | **~12.3 GB** | RTX 4080 (16GB) |
| **Gemma-4 26B-A4B** | **2 bit** | 9.26 GB | **~11.8 GB** | RTX 4070 Ti / 4080 (16GB) |
| **Gemma-3 12B** | **4 bit** | 7.80 GB | **~9.8 GB** | RTX 3080 / 4070 (12GB) |

**Gemma-3 4B** 모델 전반 및 12B 모델의 2bit 버전이 포함되며, 대부분의 메인스트림 GPU와 게이밍 노트북에서 구동 가능합니다.

```python
from unsloth.chat_templates import get_chat_template

# 1. 모델 경로/이름을 소문자로 변환하여 분석
# (만약 이어서 학습하는 모드라면 LORA_ADAPTER_NAME을 기준으로 판단할 수도 있습니다)
model_name_for_template = BASE_MODEL_ID.lower()

# 2. 버전에 맞는 템플릿 문자열 자동 할당
if "gemma-4" in model_name_for_template:
    target_template = "gemma-4"
elif "gemma-3" in model_name_for_template:
    target_template = "gemma-3"
elif "gemma" in model_name_for_template:
    # gemma-2 또는 초기 gemma 모델인 경우
    target_template = "gemma"
else:
    # Gemma 시리즈가 아닌 다른 모델을 넣었을 때의 기본 폴백 (보통 chatml을 많이 씁니다)
    target_template = "chatml"
    print("⚠️ Gemma 시리즈가 아닌 모델이 감지되었습니다. 기본 ChatML 템플릿을 시도합니다.")

print(f"🤖 모델 이름({BASE_MODEL_ID})을 분석하여 '{target_template}' 챗 템플릿을 적용합니다.")

# 3. 자동 할당된 템플릿 적용
tokenizer = get_chat_template(
    tokenizer,
    chat_template = target_template,
)
```

```python
import json
from datasets import load_dataset

ids_to_update = []

# DB에서 데이터 추출 함수
def fetch_logs_with_ids():
    import mysql.connector
    try:
        conn = mysql.connector.connect(host=MYSQL_HOST, database=MYSQL_DATABASE, user=MYSQL_USER, password=MYSQL_PASSWORD, port=MYSQL_PORT)
        if conn.is_connected():
            cursor = conn.cursor(dictionary=True)
            cursor.execute("SELECT id, user, feedbackData FROM logs WHERE isFeedback = 1 AND isLearning = 0 AND user IS NOT NULL AND feedbackData IS NOT NULL")
            results = cursor.fetchall()
            dataset, ids = [], []
            for row in results:
                dataset.append({"instruction": row["user"].strip(), "response": row["feedbackData"].strip()})
                ids.append(row["id"])
            return dataset, ids
    except Exception as e:
        print(f"MySQL Error: {e}")
    finally:
        if 'conn' in locals() and conn.is_connected():
            cursor.close(); conn.close()
    return [], []

# 데이터셋 로드
if IS_NEW_FINETUNING:
    print(f"📂 HuggingFace 데이터셋 로드: {DATASET_NAME}")
    dataset = load_dataset(DATASET_NAME, split="train")
else:
    data, ids_to_update = fetch_logs_with_ids()
    with open("lucia_dataset.json", "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=2)
    print(f"📂 로컬 DB 데이터셋 로드: 총 {len(data)}건 (업데이트 대상 ID: {len(ids_to_update)}건)")
    dataset = load_dataset("json", data_files="lucia_dataset.json", split="train")

# Unsloth 지원 템플릿 변환
def formatting_prompts_func(examples):
    conversations = []
    for inst, resp in zip(examples["instruction"], examples["response"]):
        conversations.append([
            {"content": characterConcept_detail, "role": "system"},
            {"content": inst, "role": "user"},
            {"content": resp, "role": "assistant"}
        ])
    return {"conversations": conversations}

dataset = dataset.map(formatting_prompts_func, batched=True, remove_columns=["instruction", "response"])

# 모델 전용 Chat Template 적용
def apply_chat_template(examples):
    return {"text": tokenizer.apply_chat_template(examples["conversations"], tokenize=False)}

dataset = dataset.map(apply_chat_template, batched=True)
print("✅ 데이터셋 가공 완료. 첫 번째 데이터 예시:")
print(dataset[0]["text"]) # 너무 기니까 일부만 출력
```

<a name="Train"></a>
### 🏃 모델 학습 시작 (Trainer)

```python
import os
import platform
import torch
from trl import SFTTrainer, SFTConfig
from unsloth import is_bfloat16_supported
from transformers.trainer_utils import get_last_checkpoint

# ==========================================
# [1] OS 환경 및 경로 설정
# ==========================================
is_windows = platform.system() == "Windows"
num_proc_config = None if is_windows else os.cpu_count()
num_worker_config = 0 if is_windows else 4
IS_COLAB = "COLAB_" in "".join(os.environ.keys())

# 체크포인트 저장 경로 설정 (🌟 개별화 적용 완료)
if IS_COLAB:
    # 코랩: 구글 드라이브 연결 경로 하위에 어댑터명으로 개별 폴더 생성
    output_dir = f"/content/drive/MyDrive/Lucia_Checkpoints/{LORA_ADAPTER_NAME}"
else:
    # 로컬: 현재 폴더 내에 모델명을 딴 체크포인트 폴더 생성
    output_dir = f"./{LORA_ADAPTER_NAME}_checkpoints"

os.makedirs(output_dir, exist_ok=True)

# 기존에 학습하던 '일기장(체크포인트)'이 있는지 확인
last_checkpoint = get_last_checkpoint(output_dir)

if last_checkpoint is not None:
    print(f"👀 발견된 체크포인트: ({last_checkpoint})")
    print("🔄 해당 지점부터 학습을 이어서 시작합니다.")
else:
    print(f"🌱 {output_dir} 폴더에 기존 기록이 없습니다. 처음부터 학습을 시작합니다.")

print(f"⚙️ 학습 환경: OS={platform.system()}, 멀티프로세싱={num_proc_config}")

# ==========================================
# [2] SFTTrainer 설정 (로컬/코랩 통합)
# ==========================================
# 에러 방지를 위해 packing은 False로 고정합니다.
packing_config = False

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    eval_dataset = None,
    args = SFTConfig(
        dataset_text_field = "text",
        max_seq_length = max_seq_length,

        # --- 학습 효율 및 안정성 설정 ---
        per_device_train_batch_size = 2 if IS_COLAB else 1,
        gradient_accumulation_steps = 8 if IS_COLAB else 4,

        # --- 학습량 및 하이퍼파라미터 ---
        num_train_epochs = 3,
        max_steps = -1,
        learning_rate = 1e-4 if IS_COLAB else 2e-5,
        max_grad_norm = 0.3,

        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "cosine",
        warmup_ratio = 0.1,

        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),

        logging_steps = 10,
        seed = 3407,
        report_to = "none",

        # --- 체크포인트 저장 전략 (중요) ---
        output_dir = output_dir,
        save_strategy = "epoch",    # 에포크마다 저장 (또는 "steps"로 변경 가능)
        save_total_limit = 3,       # 용량 관리를 위해 최신 3개만 유지

        # --- 환경 설정 적용 ---
        packing = packing_config,
        dataset_num_proc = num_proc_config,
        dataloader_num_workers = num_worker_config
    ),
)

# ==========================================
# [3] 파인튜닝 실행
# ==========================================
print(f"🚀 {'Gemma 4' if 'gemma-4' in BASE_MODEL_ID.lower() else 'Gemma 3'} 최적화 학습을 시작합니다...")

# resume_from_checkpoint 인자를 통해 중단 지점부터 이어서 학습
trainer_stats = trainer.train(resume_from_checkpoint = last_checkpoint)

print("✅ 모든 학습 프로세스가 완료되었습니다!")
```

### 📉 ExLlamaV3 양자화 진행 (진행 도중 에러 발생시 당황하지마시고 이부분부터 다시 시작하시길 바랍니다.)

```python
import os
import shutil
import site

# ==============================================================
# [1] 모델 경로 및 환경 설정
# ==============================================================
log_file_path = "last_saved_model_name.txt"
if os.path.exists(log_file_path):
    with open(log_file_path, "r", encoding="utf-8") as f:
        input_model = f.read().strip()
    print(f"📂 병합 모델 경로 감지됨: {input_model}")
else:
    input_model = f"{LORA_ADAPTER_NAME}_gemma-4"
    print(f"⚠️ 기록 파일이 없어 기본 경로를 사용합니다: {input_model}")

IS_COLAB = "COLAB_" in "".join(os.environ.keys())

if IS_COLAB:
    if not os.path.exists("exllamav3"):
        print("📥 exllamav3 리포지토리를 클론합니다...")
        !git clone https://github.com/turboderp-org/exllamav3

    # ==============================================================
    # [2] Colab 전용: formatron Pydantic 의존성 패치
    # ==============================================================
    patched = False
    for site_pkg in site.getsitepackages():
        formatron_dir = os.path.join(site_pkg, "formatron")
        if os.path.exists(formatron_dir):
            for root, dirs, files in os.walk(formatron_dir):
                for file in files:
                    if file.endswith(".py"):
                        file_path = os.path.join(root, file)
                        with open(file_path, "r", encoding="utf-8") as f:
                            code = f.read()

                        original_code = code
                        code = code.replace("from pydantic import typing", "import typing")
                        code = code.replace("import pydantic.typing as typing", "import typing")
                        code = code.replace("pydantic.typing", "typing")

                        if code != original_code:
                            with open(file_path, "w", encoding="utf-8") as f:
                                f.write(code)
                            patched = True
            if patched:
                print("✅ formatron 라이브러리 Pydantic 패치 완료!")
            break

# ==============================================================
# [3] ExLlamaV3 양자화 및 즉시 업로드 (용량 극한 절약 루프)
# ==============================================================
base_dir = f"workspace/{input_model}_exllama3"
target_bits = [8.0, 6.0, 4.0, 2.0]

if IS_COLAB and globals().get("UPLOAD_TO_HF", False):
    from huggingface_hub import HfApi
    api = HfApi()

# 본격적인 양자화 루프 시작
for bits in target_bits:
    out_dir = f"{base_dir}/ExtResult_{bits}bpw"
    work_dir = f"{base_dir}/Working_{bits}bpw"

    if os.path.exists(os.path.join(out_dir, "config.json")):
        print(f"\n✅ {bits}bpw 변환은 이미 완료되어 건너뜁니다.")
        continue

    print("\n" + "="*40)
    resume_cmd = "-resume" if os.path.exists(work_dir) else ""
    print(f"🚀 {bits}bpw 양자화 시작! (이어하기 여부: {bool(resume_cmd)})")
    print("="*40)

    # 양자화 실행
    !python exllamav3/convert.py -i {input_model} -o {out_dir} -w {work_dir} -b {bits} -cpi 0 -d 0 {resume_cmd} -cr 100 -cc 8096

    if os.path.exists(os.path.join(out_dir, "config.json")):
        print(f"✅ {bits}bpw 변환 성공!")

        # 1. 공통 처리: Working 폴더 즉시 삭제 (임시 데이터 제거)
        if os.path.exists(work_dir):
            shutil.rmtree(work_dir)
            print(f"🗑️ 임시 작업 폴더 삭제 완료: {work_dir}")

        # 2. Colab 전용 처리: 개별 레포지토리 생성 및 Root 경로 업로드
        if IS_COLAB and globals().get("UPLOAD_TO_HF", False):
            # 양자화 비트(bits)를 레포지토리 이름 끝에 추가하여 독립된 레포 생성
            auto_repo_id = f"{HF_USER_ID}/{input_model}_exllama3_{bits}bpw"
            print(f"\n🚀 Hugging Face Hub 업로드 준비: {auto_repo_id}")

            try:
                # 레포지토리 생성 (이미 있으면 무시)
                api.create_repo(repo_id=auto_repo_id, token=HF_TOKEN, exist_ok=True, private=False)

                print(f"📤 {bits}bpw 모델 최상위 경로(Root)에 업로드 중...")
                # path_in_repo 파라미터를 삭제하면 폴더 내부의 파일들이 레포지토리 최상위에 바로 업로드됩니다.
                api.upload_folder(
                    folder_path=out_dir,
                    repo_id=auto_repo_id,
                    repo_type="model",
                    token=HF_TOKEN
                )
                print(f"✅ {bits}bpw 업로드 완료! (https://huggingface.co/{auto_repo_id})")

                # 업로드가 끝났으므로 코랩 디스크에서 결과물 완전 삭제
                shutil.rmtree(out_dir)
                print(f"🔥 코랩 용량 확보를 위해 결과 폴더 완전 삭제 완료: {out_dir}")

            except Exception as e:
                print(f"⚠️ 업로드 중 오류 발생 (안전을 위해 폴더는 삭제하지 않습니다): {e}")

    else:
        print(f"⚠️ {bits}bpw 변환이 중단되었습니다. 다음 과정으로 넘어가지 않습니다.")
        break

print("\n🎉 모든 양자화 및 업로드 과정 종료!")
```
