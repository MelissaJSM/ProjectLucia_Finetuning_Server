# 📊 LLM 모델별 용량 및 필요 VRAM 추정표

> **참고:**
> * **Lock / Unlock:** 두 버전의 용량은 동일하므로 하나로 통합하여 표기했습니다.
> * **추정 VRAM:** 모델 가중치(파일 크기) + `2~3GB` (KV Cache 및 시스템 오버헤드)를 포함한 수치입니다. 컨텍스트 길이(Context Length)를 길게 설정하면 더 많은 메모리가 필요합니다.

---

## 1. High-End (24GB VRAM 이상 권장)
**Gemma-3 27B**는 고성능 모델로, 4bit 양자화조차 16GB 카드에서는 구동이 어렵습니다.

| 모델 (Model) | 양자화 (Bit) | 파일 용량 | 추정 필요 VRAM | 권장 GPU |
| :--- | :---: | :---: | :---: | :--- |
| **Gemma-3 27B** | **8 bit** | 27.50 GB | **~30.5 GB** | A6000 / 3090 NVLink |
| **Gemma-3 27B** | **6 bit** | 21.54 GB | **~24.5 GB** | RTX 3090 / 4090 (여유 없음) |
| **Gemma-3 27B** | **4 bit** | 15.58 GB | **~18.6 GB** | RTX 3090 / 4090 |
| **Gemma-3 27B** | **2 bit** | 9.62 GB | **~12.6 GB** | RTX 4080 (16GB) |

---

## 2. Sweet Spot (12GB ~ 16GB VRAM 권장)
**Phi-4 (14B)**와 **Gemma-3 12B**는 일반적인 고사양 게이밍 GPU에서 가장 효율이 좋은 구간입니다.

| 모델 (Model) | 양자화 (Bit) | 파일 용량 | 추정 필요 VRAM | 권장 GPU |
| :--- | :---: | :---: | :---: | :--- |
| **Phi-4 (14B)** | **8 bit** | 13.80 GB | **~15.8 GB** | RTX 4080 / 4070 Ti Super |
| **Gemma-3 12B** | **8 bit** | 12.81 GB | **~14.8 GB** | RTX 4080 / 4070 Ti Super |
| **Phi-4 (14B)** | **6 bit** | 10.63 GB | **~12.6 GB** | RTX 4080 (16GB) |
| **Gemma-3 12B** | **6 bit** | 10.31 GB | **~12.3 GB** | RTX 4080 (16GB) |
| **Gemma-3 12B** | **4 bit** | 7.80 GB | **~9.8 GB** | RTX 3080 (10GB↑) |
| **Phi-4 (14B)** | **4 bit** | 7.46 GB | **~9.5 GB** | RTX 3080 (10GB↑) |
| **Gemma-3 12B** | **2 bit** | 5.29 GB | **~7.3 GB** | RTX 3060 / 4060 (8GB) |
| **Phi-4 (14B)** | **2 bit** | 4.28 GB | **~6.3 GB** | RTX 3060 / 4060 (8GB) |

---

## 3. Light Weight (6GB ~ 8GB VRAM 권장)
**Gemma-3 4B**는 대부분의 메인스트림 GPU와 게이밍 노트북에서 구동 가능합니다.

| 모델 (Model) | 양자화 (Bit) | 파일 용량 | 추정 필요 VRAM | 권장 GPU |
| :--- | :---: | :---: | :---: | :--- |
| **Gemma-3 4B** | **8 bit** | 5.06 GB | **~6.6 GB** | RTX 3060 / 4060 (8GB) |
| **Gemma-3 4B** | **6 bit** | 4.31 GB | **~5.8 GB** | RTX 2060 / 3060 (6GB↑) |
| **Gemma-3 4B** | **4 bit** | 3.56 GB | **~5.1 GB** | GTX 1660S / 3050 (6GB) |
| **Gemma-3 4B** | **2 bit** | 2.81 GB | **~4.3 GB** | GTX 1660S / 3050 (6GB) |

---

## 4. Ultra Light (저사양 / 내장 그래픽 가능)
**1B**와 **270M**은 VRAM 4GB 미만 환경이나 CPU 모드에서도 빠르게 작동합니다.

| 모델 (Model) | 양자화 (Bit) | 파일 용량 | 추정 필요 VRAM | 비고 |
| :--- | :---: | :---: | :---: | :--- |

```python
# 추가 파인튜닝을 위한 대화 피드백 데이터셋 작업 (DB 연동)
if not IS_NEW_FINETUNING:
    import mysql.connector
    from mysql.connector import Error

    def fetch_logs_with_ids():
        try:
            connection = mysql.connector.connect(
                host=MYSQL_HOST,
                database=MYSQL_DATABASE,
                user=MYSQL_USER,
                password=MYSQL_PASSWORD,
                port=MYSQL_PORT
            )

            if connection.is_connected():
                cursor = connection.cursor(dictionary=True)
                query = """
                    SELECT id, user, feedbackData
                    FROM logs
                    WHERE isFeedback = 1 AND isLearning = 0
                    AND user IS NOT NULL AND feedbackData IS NOT NULL
                """
                cursor.execute(query)
                results = cursor.fetchall()

                dataset = []
                ids_to_update = []

                for row in results:
                    dataset.append({
                        "instruction": row["user"].strip(),
                        "response": row["feedbackData"].strip()
                    })
                    ids_to_update.append(row["id"])

                return dataset, ids_to_update

        except Error as e:
            print(f"MySQL Error: {e}")
            return [], []

        finally:
            if 'connection' in locals() and connection.is_connected():
                cursor.close()
                connection.close()
else:
    print("✅ 초기 학습 모드이므로 DB 연결을 건너뜁니다.")
```

```python
import platform
from trl import SFTTrainer, SFTConfig

# [설정] 현재 OS가 윈도우인지 확인
is_windows = platform.system() == "Windows"

# [핵심 수정] 윈도우에서는 1이 아니라 'None'을 줘야 별도 프로세스를 안 만듭니다!
# 윈도우: None (메인 프로세스에서 실행)
# 리눅스: None (자동) 또는 숫자 지정
num_proc_config = None
num_worker_config = 0 if is_windows else 4
packing_config = False if is_windows else True

print(f"🖥️ 감지된 OS: {platform.system()}")
print(f"⚙️ 병렬 처리 설정: num_proc={num_proc_config} (None=Main Process), num_workers={num_worker_config}")

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    eval_dataset = None,
    args = SFTConfig(
        dataset_text_field = "text",
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        num_train_epochs = 3, # 학습 데이터셋 사이클 설정입니다. 스탭을 쓴다면 0으로 설정해주세요.
        max_steps = -1, # 학습 스탭 설정입니다. 에포크를 쓴다면 -1 로 설정해주세요
        learning_rate = 2e-5,
        logging_steps = 100,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        report_to = "none",

        # [중요] 윈도우 에러 방지 핵심 설정
        packing = packing_config,            # 패킹 끄기 (충돌 원인 제거)
        dataset_num_proc = num_proc_config, # None이어야 메인 프로세스에서 돕니다
        dataloader_num_workers = num_worker_config
    ),
)
```

```python
trainer_stats = trainer.train()
```
